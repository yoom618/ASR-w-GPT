# LIBRI_SPEECH

- `split = 'train'`
    - `subconfig="clean.100"`
    - `subconfig="clean.360"`
    - `subconfig="clean.500"`
    - default : all
- `split = 'validation.clean'`
- `split = 'validation.other'`
- `split = 'test.clean'`
- `split = 'test.other'`


```python
import os
from datasets import load_dataset
from pprint import pprint
from IPython.display import display, Audio

os.environ['HF_HOME'] = "/data/yoom618/datasets"

import aiohttp

librispeech = load_dataset("openslr/librispeech_asr", 
                           "all", # all, clean, other
                           cache_dir="/data/yoom618/datasets",
                           trust_remote_code=True,
                           storage_options={'client_kwargs': {'timeout': aiohttp.ClientTimeout(total=7200)}}
                          )

print(librispeech)

display(Audio(librispeech['train.clean.100'][0]['audio']['array'], 
              rate=librispeech['train.clean.100'][0]['audio']['sampling_rate'], autoplay=False))
pprint(librispeech['train.clean.100'][0])

del librispeech
```
